```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Instructions/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "Instructions/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
school_data_complete.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#District Summary Calculations

#calculate the total number of schools
school_count = len(school_data_complete['School ID'].unique())
#school_count
```

```python
#calculate the total number of students
student_count = school_data_complete['Student ID'].count()
#student_count
```

```python
#Total budget for each school
budget_total = school_data['budget'].sum()
#budget_total
```

```python
#calculate the average math score
avg_math = school_data_complete['math_score'].mean()
#avg_math
```

```python
#calculate the average reading score
avg_reading = school_data_complete['reading_score'].mean()
#avg_reading
```

```python
#calculate the percentage of students with a passing math score
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]

#calculate the total number of students with a passing math score
passing_math_count = passing_math['Student ID'].count()

#calculate the pct of students with a passing math score
math_pct = passing_math_count/student_count
#math_pct
```

```python
#extract the number of students with a passing reading score
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]

#count up the total number of students with a passing reading score
passing_reading_count = passing_reading['Student ID'].count()
                                    
#calculate the pct of students with a passing reading score
reading_pct = passing_reading_count/student_count
#reading_pct
```

```python
#calculate the percentage of students with a passing math and reading score
passing_both = school_data_complete.loc[(school_data_complete['math_score']>= 70) & (school_data_complete['reading_score'] >= 70)]

#calculate the total number of students with a passing math and reading score
passing_both_count = passing_both['Student ID'].count()

#calculate the pct of students with a passing math and reading score
both_pct = passing_both_count/student_count 
#both_pct
```

```python
#create a new DataFrame
district_summary_df = pd.DataFrame({"Total Schools":[school_count],
                     "Total Students":[student_count],
                     "Total Budget":[budget_total],
                     "Average Math Score":[avg_math],
                     "Average Reading Score":[avg_reading],
                     "% Passing Math":[math_pct],
                     "% Passing Reading":[reading_pct],
                     "% Overall Passing":[both_pct]})
```

```python
# Use Map to format all the columns
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("{:,}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:,.2f}".format)
district_summary_df["% Passing Math"] = (district_summary_df["% Passing Math"]*100).map("{:.1f}%".format)
district_summary_df["% Passing Reading"] = (district_summary_df["% Passing Reading"]*100).map("{:.1f}%".format)
district_summary_df["% Overall Passing"] = (district_summary_df["% Overall Passing"]*100).map("{:.1f}%".format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#group by each school id and complete specific calculations to the columns
school_df = school_data_complete.groupby(['School ID']).agg({'school_name':'first', 'type':'first',
                                                            'Student ID':'count', 'budget':'first',
                                                            'math_score':'mean', 'reading_score':'mean', 
                                                            })
#school_df
```

```python
#calculate the percentage of students with a passing math score
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]
pass_math_per_school = passing_math.groupby(["School ID"]).count()["Student ID"]
pct_math = pass_math_per_school/school_df["Student ID"] * 100
```

```python
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
pass_reading_per_school = passing_reading.groupby(["School ID"]).count()["Student ID"]
pct_read = pass_reading_per_school/school_df["Student ID"] * 100
```

```python
#calculate the percentage of students with a passing math and reading score
passing_both = school_data_complete.loc[(school_data_complete['math_score']>= 70) & (school_data_complete['reading_score'] >= 70)]
pass_both_per_school = passing_both.groupby(["School ID"]).count()["Student ID"]
pct_both = pass_both_per_school/school_df["Student ID"] * 100
```

```python
#create a DataFrame within a dictionary
merge_data = pd.DataFrame({'% Passing Math':pct_math,'% Passing Reading':pct_read,'% Overall Passing':pct_both})
#merge_data
```

```python
school_df = school_data_complete.groupby(['School ID']).agg({'school_name':'first', 'type':'first',
                                                            'Student ID':'count', 'budget':'first',
                                                            'math_score':'mean', 'reading_score':'mean', 
                                                            })
#school_df
```

```python
school_summary_df = pd.merge(school_df, merge_data, on="School ID", how="outer")
#school_summary_df
```

```python
#per student budget
per_student_budget = school_summary_df['budget']/school_summary_df['Student ID']
#per_student_budget
```

```python
merge_budget = pd.DataFrame({'Per Student Budget':per_student_budget})
#merge_budget
```

```python
school_summary_df2 = pd.merge(school_summary_df, merge_budget, on="School ID", how="outer")
#school_summary_df2.columns
```

```python
school_summary_renamed = school_summary_df2.rename(columns = {'school_name':"", 'type':"School Type", 
                                                              'Student ID':"Total Students", 
                                                              'budget':"Total School Budget", 
                                                              'math_score':"Average Math Score",
                                                              'reading_score':"Average Reading Score",})  
school_summary_final = school_summary_renamed[["",'School Type', 'Total Students', 'Total School Budget',
       'Per Student Budget','Average Math Score', 'Average Reading Score', '% Passing Math',
       '% Passing Reading', '% Overall Passing']]
#school_summary_final
```

```python
school_summary_final["Total School Budget"] = school_summary_final["Total School Budget"].map("{:,}".format)
school_summary_final["Per Student Budget"] = school_summary_final["Per Student Budget"].map("${:.1f}".format)
school_summary_final["Average Math Score"] = school_summary_final["Average Math Score"].map("{:,.2f}".format)
school_summary_final["Average Reading Score"] = school_summary_final["Average Reading Score"].map("{:,.2f}".format)
school_summary_final["% Passing Math"] = (school_summary_final["% Passing Math"]).map("{:.1f}%".format)
school_summary_final["% Passing Reading"] = (school_summary_final["% Passing Reading"]).map("{:.1f}%".format)
school_summary_final["% Overall Passing"] = (school_summary_final["% Overall Passing"]).map("{:.1f}%".format)
school_summary_final
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth_grade_math = school_data_complete[school_data_complete["grade"]=="9th"].groupby("school_name").mean()["math_score"]
tenth_grade_math = school_data_complete[school_data_complete["grade"]=="10th"].groupby("school_name").mean()["math_score"]
eleventh_grade_math = school_data_complete[school_data_complete["grade"]=="11th"].groupby("school_name").mean()["math_score"]
twelfth_grade_math = school_data_complete[school_data_complete["grade"]=="12th"].groupby("school_name").mean()["math_score"]

math_scores_grade = pd.DataFrame({"9th Grade":ninth_grade_math,"10th Grade":tenth_grade_math,"11th Grade":eleventh_grade_math,"12th Grade": twelfth_grade_math})

math_scores_grade["9th Grade"] = math_scores_grade["9th Grade"].map("{:,.2f}".format)
math_scores_grade["10th Grade"] = math_scores_grade["10th Grade"].map("{:,.2f}".format)
math_scores_grade["11th Grade"] = math_scores_grade["11th Grade"].map("{:,.2f}".format)
math_scores_grade["12th Grade"] = math_scores_grade["12th Grade"].map("{:,.2f}".format)
math_scores_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Establish bins and spending labels
bins = [0,584.99,629.99,644.99,675.99]
group_names = ["<$585", "$585-630","$630-645","$645-680"]
```

```python
#Take the $ sign out of the budget by getting the dataframe before formatting
school_spending_df = school_summary_renamed

#use pd.cut to sort by binning
school_spending_df["Spending Ranges (Per Student)"]=pd.cut(school_spending_df["Per Student Budget"],bins,labels=group_names,include_lowest=True)                                                                                          


#get rid of the things you do not want
school_spending_df = school_spending_df.loc[:,['Spending Ranges (Per Student)','Average Math Score','Average Reading Score',
'% Passing Math','% Passing Reading','% Overall Passing']]

#group by Spending Ranges
scores_by_school_spending = school_spending_df.groupby('Spending Ranges (Per Student)').mean()

scores_by_school_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Establish bins and labels
school_bins = [0, 1000, 2000, 5000]
size_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_size_df = school_summary_renamed

#"Total Students an integer?"
school_size_df["Total Students"]=pd.to_numeric(school_size_df["Total Students"])
school_size_df.columns 

#use pd.cut to sort by binning #is "Total Students an integer?"
school_size_df["School Size"]=pd.cut(school_size_df['Total Students'],school_bins,labels=size_names) 

#get rid of things you do not want
school_size_df = school_size_df.loc[:,['Average Math Score','Average Reading Score',
'% Passing Math','% Passing Reading','% Overall Passing',"School Size"]]
school_size_df

#group by school size ranges
size_by_school = school_size_df.groupby("School Size").mean()

size_by_school
```
